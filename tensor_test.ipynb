# Import

```python
import numpy as np
import tensorflow
from tensorflow import keras
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

# inputs

```python
# inputs and lables
x = np.array([np.arange(1000), np.arange(1000)]).transpose()
y = (np.arange(1000) * 2.2) + 110*np.random.rand(1000)

x, y = shuffle(x, y, random_state=0)
```

# Layers

```python
train_x, train_y = x[:800], y[:800]
test_x, test_y = x[800:], y[800:]
```

```python
model = keras.Sequential() # defination
model.add(keras.layers.Dense(8, activation='relu', input_shape=(2,)))   # first layer
model.add(keras.layers.Dense(4, activation='relu'))                     # second layer
model.add(keras.layers.Dense(1, activation='linear'))                   # output

model.summary()
```

# Compile and fit

```python
  model.compile(optimizer='adam', loss='mse', metrics='mae')    # build model with loss function and optimizer  'sgd' for gradient descent
  history = model.fit(train_x, train_y, epochs=150, batch_size=32, validation_split=0.1)  # fit model (with number of epochs and bached size 
  #and validation_split for percent of data for validitaion and verbose for illustrate resulte or not)
```

# predict and plot

```python
pred = model.predict(test_x)   # predict result with trained model
```

```python
plt.scatter(range(len(test_y)), test_y)
plt.scatter(range(len(pred)), pred)
```
